```python
import os
from functools import partial

def flatten(t):
    return [item for sublist in t for item in sublist if not os.path.isdir()]

def dl_imgs(birds_dict, folder="images"):
    
    folders = [f for f in list(birds_dict.values())]
    to_dl = [item for sublist in folders for item in sublist if not os.path.isdir(folder + "/" + item)]
    
    dl = partial(downloader.download, limit=150, output_dir='images', adult_filter_off=True, force_replace=False, timeout=10)
    
    for b in to_dl:
        print("Downloading " + b)
        dl(b)
    
    print("Downloads completed.")
```
